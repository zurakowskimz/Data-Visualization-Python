```python
import numpy as np
import matplotlib
matplotlib.__version__
import matplotlib.pyplot as plt
import seaborn as sns
sns.__version__
import pandas as pd
```

```python
cmap = matplotlib.cm.get_cmap('coolwarm')
p = [0.2, 0.1, 0.4,  0.15, 0.15]
e = ["yes", "rather yes", "rather no", "no", "I don't know"]
s = 0.9  ## bar width
x = np.arange(len(p))
kolory = np.round(np.linspace(0.1,1,len(p)),1)


plt.bar(x, p, s, color = cmap(kolory))
plt.xticks(x,e, rotation = 30)
plt.show()
```

```python
plt.pie(p, colors=cmap(kolory))
plt.axis("equal")
plt.legend(e)
plt.title("Would you buy a car for 200 000 PLN?")
plt.show()
```

```python
import matplotlib.mlab as mlab
mu, sigma = 100, 15
x = mu + sigma*np.random.randn(1000)

n, bins, patches = plt.hist(x, 35, normed=1, facecolor='green', alpha=0.75, edgecolor="k")
plt.xlabel('Smarts')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
plt.axis([45, 155, 0, 0.03])
plt.grid(True)

plt.show()
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
flights_pivot = flights.pivot("month", "year", "passengers")
```

```python
sns.heatmap(flights_pivot, annot = True, fmt = "d", linewidths=0.8)
plt.show()
```

```python
averagePassengers = flights["passengers"].groupby(flights["year"]).mean()
```

```python
ax = plt.subplot()
ax.plot(averagePassengers.index, averagePassengers.values, "bp--", label = "Average number of passengers")
l = ax.legend()
for text in l.get_texts():
    text.set_color("red")
ax.tick_params(axis='x', colors = "green")
ax.tick_params(axis='y', colors = "green")
ax.set_title("Chart showing average number of passengers", color="magenta")
ax.axhline(y=240, color='k', linestyle="--")
ax.axvline(x=1958, color='k', linestyle="--")
ax.set_facecolor("white")
ax.set_xlabel("Year", color = "red")
#ax.grid(True, which='both')
plt.show()
```

```python
import datetime
teraz = datetime.datetime.now()
theta = np.linspace(0,np.pi*2,100)
y = 0+np.sin(theta)
x = 0+np.cos(theta)
plt.fill(x,y,fill=True,color="k", alpha=0.1)
plt.plot(x,y,'k')
plt.axis("equal")
plt.text(0, -0.9, "6", fontsize=20)
plt.text(0, 0.8, "12", fontsize=20)
plt.text(0.8, 0.1, "3", fontsize=20)
plt.text(-0.9, 0.1, "9", fontsize=20)
x3 = np.linspace(0.75,1,2)
y3 = [0,0]
plt.plot(x3,y3, color="k")
x9 = np.linspace(-1,-0.75,2)
plt.plot(x9,y3,color="k")
y12 = np.linspace(0.75,1,2)
x12 = [0,0]
y6 = np.linspace(-1,-0.75,2)
plt.plot(x12,y12,color="k")
plt.plot(x12,y6,color="k")
plt.scatter(0,0,color="k")

## wskazwki
godzina = (teraz.hour % 12) / 12
godzina_obrot = (0.5*np.pi) - godzina*(2*np.pi)
x_godzina = np.cos(godzina_obrot)
y_godzina = np.sin(godzina_obrot)
plt.plot([0,0.5*x_godzina],[0,0.5*y_godzina])


minuta = (teraz.minute /60)
minuta_obrot = (0.5*np.pi) - minuta*(2*np.pi)
x_minuta = np.cos(minuta_obrot)
y_minuta = np.sin(minuta_obrot)
plt.plot([0,0.75*x_minuta],[0,0.75*y_minuta])
plt.axis("off")
plt.show()
```
